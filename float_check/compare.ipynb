```python
import numpy as np
import pandas as pd
import sqlite3
```

```python
restore1 = np.load('run_on_osx/test_data.npz')
restore2 = np.load('run_on_hyak/test_data.npz')
```

```python
for key in restore1.keys():
    diff = restore1[key] - restore2[key]
    print(restore1[key], restore2[key], diff)
    if np.max(np.abs(diff)) != 0:
        print('There is a difference')
```

```python
conn1 = sqlite3.connect('run_on_osx/test_data.sqlite')
conn2 = sqlite3.connect('run_on_hyak/test_data.sqlite')

df1 = pd.read_sql('select * from num_floats;', conn1)
df2 = pd.read_sql('select * from num_floats;', conn2)
```
