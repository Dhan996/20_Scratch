```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp

import lsst.sims.maf.db as db
import lsst.sims.maf.utils as utils
import lsst.sims.maf.metrics as metrics
import lsst.sims.maf.slicers as slicers
import lsst.sims.maf.stackers as stackers
import lsst.sims.maf.metricBundles as metricBundles

import glob
```

```python
class NafterNMetric(metrics.BaseMetric):
    """Night after N metrics are taken
    """
    def __init__(self, nightCol='night', n_needed=1, metricName='NafterN', badVal=-666, **kwargs):
        self.n_needed = n_needed
        self.nightCol = nightCol
        self.badVal = badVal
        super().__init__(col=[nightCol], metricName=metricName, units='night after %i' % n_needed, **kwargs)
    def run(self, dataSlice, slicePoint=None):
        dataSlice.sort(order=self.nightCol)
        if np.size(dataSlice) >= self.n_needed:
            result = dataSlice[self.nightCol][self.n_needed -1]
        else:
            result = self.badVal
        return result
    
```

```python
dbfile = 'baseline_nexp2_v1.7_10yrs.db'
outDir='temp'
resultsDb = db.ResultsDb(outDir=outDir)
conn = db.OpsimDatabase(dbfile)
```

```python
for n_needed in [1, 3, 5]:
    bundleList = []
    for filtername in 'ugrizy':
        sql = 'night < 730 and filter="%s"' % filtername
        metric = NafterNMetric(nightCol='night', n_needed=n_needed, metricName='MinNight_%s_%i' % (filtername, n_needed))
        slicer = slicers.HealpixSlicer()
        summaryStats = [metrics.SumMetric()]
        plotDict = {'colorMax': 365, 'cumulative': True}
        bundleList.append(metricBundles.MetricBundle(metric,slicer,sql, plotDict=plotDict, summaryMetrics=summaryStats))
    bd = metricBundles.makeBundlesDictFromList(bundleList)
    bg = metricBundles.MetricBundleGroup(bd, conn, outDir=outDir, resultsDb=resultsDb)
    bg.runAll()
    bg.plotAll(closefigs=False)
```
